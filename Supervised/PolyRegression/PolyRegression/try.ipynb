```python
import numpy as np  
import matplotlib.pyplot as plt    
from sklearn.preprocessing import PolynomialFeatures  
from sklearn.linear_model import LinearRegression  
```

```python
np.random.seed(0) 
X=2*np.random.rand(100,1) 
y=3*X**2 + 2*X + 1 + np.random.randn(100,1) 
```

```python
plt.scatter(X,y,color="green",label="Real Data") 
plt.title("Dataset with Quadratic Equation") 
plt.xlabel("X") 
plt.ylabel("y") 
plt.legend() 
plt.show()
```

```python
poly_feature = PolynomialFeatures(degree=2) 
X_poly= poly_feature.fit_transform(X)
```

```python
model = LinearRegression() 
model.fit(X_poly,y)
```

```python
X_new = np.linspace(0,2,100).reshape(-1,1)
X_new_poly = poly_feature.transform(X_new)
y_new = model.predict(X_new_poly)
```

```python
plt.scatter(X,y,color="pink", label="Real Data")
plt.plot(X_new,y_new,color="red", label="Poly Data")
plt.title("Poly Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
